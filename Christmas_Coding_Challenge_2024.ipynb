# Day 1

Today I have created an artificial data set using python.

```python
import pandas as pd
import numpy as np
from datetime import datetime, timedelta
```

```python
# Define date range
start_date = datetime(2024, 12, 1)
end_date = datetime(2024, 12, 31)
date_range = pd.date_range(start=start_date, end=end_date, freq='D')

# Generate dummy data
data = {
    "Date": date_range,
    "Heart Rate (bpm)": np.random.randint(60, 120, size=len(date_range)),
    "Steps": np.random.randint(2000, 15000, size=len(date_range)),
    "Calories Burned (kcal)": np.random.randint(1500, 3500, size=len(date_range)),
    "Active Hours (hrs)": np.random.uniform(1, 8, size=len(date_range)).round(2),
    "Stand Hours (hrs)": np.random.randint(8, 16, size=len(date_range)),
    "Sleep Hours (hrs)": np.random.uniform(5, 9, size=len(date_range)).round(2),
    "Activity Ring Completion (%)": np.random.randint(50, 100, size=len(date_range)),
}

# Create DataFrame
df = pd.DataFrame(data)
```

```python
import matplotlib.pyplot as plt
```

```python
df.plot(x='Date', y='Calories Burned (kcal)')
plt.title('Calories Burned Over Time')
plt.xlabel('Date')
plt.ylabel('Calories Burned (kcal)')
plt.show()
```

```python
df.plot(x='Date', y='Activity Ring Completion (%)')
plt.title('Activity Ring Completion Over Time')
plt.xlabel('Date')
plt.ylabel('Activity Ring Completion (%)')
plt.show()
```
