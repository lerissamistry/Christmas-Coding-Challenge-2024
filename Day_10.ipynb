**Problem: Fitness Tracker Data Analysis**

You have a CSV file containing daily fitness tracker data. Each row represents a day's activity and contains the following columns:

Date: The date of the activity.
Steps: Total steps walked that day.
Calories: Total calories burned.
Minutes_Active: Total active minutes.
Distance: Total distance walked (in kilometers).

```python
import pandas as pd
import numpy as np
```

```python
data = {
    "Date": ["2024-12-01", "2024-12-02", "2024-12-03", "2024-12-04",
             "2024-12-05", "2024-12-06", "2024-12-07"],
    "Steps": [5600, 9800, 7200, 6200, 10500, 8600, 4000],
    "Calories": [250, 400, 310, 280, 450, 390, 180],
    "Minutes_Active": [50, 80, 65, 55, 90, 75, 35],
    "Distance": [4.5, 7.2, 5.8, 5.0, 8.0, 6.8, 3.2]
}

# Create a DataFrame
df = pd.DataFrame(data)

# Convert Date column to datetime type for better handling
df["Date"] = pd.to_datetime(df["Date"])
```

```python
# Weekly Summary Calculations
total_steps = df["Steps"].sum()
average_steps = df["Steps"].mean()

total_calories = df["Calories"].sum()
average_calories = df["Calories"].mean()

total_active_minutes = df["Minutes_Active"].sum()
average_active_minutes = df["Minutes_Active"].mean()

total_distance = df["Distance"].sum()
average_distance = df["Distance"].mean()
```

```python
# Find day with highest and lowest steps
day_highest_steps = df.loc[df["Steps"].idxmax()]
day_lowest_steps = df.loc[df["Steps"].idxmin()]

# Filter days based on conditions
steps_above_8000 = df[df["Steps"] > 8000]
high_calories_active_minutes = df[(df["Calories"] > 350) & (df["Minutes_Active"] > 70)]
```

```python
# Display results
print("Weekly Summary:")
print(f"Total Steps: {total_steps}")
print(f"Average Steps: {average_steps:.2f}")
print(f"Total Calories Burned: {total_calories}")
print(f"Average Calories Burned: {average_calories:.2f}")
print(f"Total Active Minutes: {total_active_minutes}")
print(f"Average Active Minutes: {average_active_minutes:.2f}")
print(f"Total Distance: {total_distance:.2f} km")
print(f"Average Distance: {average_distance:.2f} km\n")

print(f"Day with the highest steps: {day_highest_steps['Date'].strftime('%Y-%m-%d')} ({day_highest_steps['Steps']} steps)")
print(f"Day with the lowest steps: {day_lowest_steps['Date'].strftime('%Y-%m-%d')} ({day_lowest_steps['Steps']} steps)\n")

print("Days with steps > 8000:")
print(steps_above_8000, "\n")

print("Days with calories burned > 350 and active minutes > 70:")
print(high_calories_active_minutes)
```
